```python
import pandas as pd
import numpy as np
import hashlib 
```

```python
def get_hash(i):
    '''Hashing the value i'''
    return hashlib.md5(str.encode(str(i))).hexdigest()

def get_values(lst, size, p=None):
    '''Obtaining a list of values, taking into account the probability of occurrence of elements'''
    return [np.random.choice(lst, size=1, replace=True, p=p)[0] for i in range(0, size)]
```

```python
def get_GP(size):
    '''Generated date set with given size of users'''
    user_ids = [get_hash(i) for i in range(0, size)]
    countries = get_values(lst=['US', 'DE', 'ZW'], p=[0.6, 0.3, 0.1], size=len(user_ids))
    devices = get_values(lst=['android', 'ios'], p=[0.7, 0.3], size=len(user_ids))
    partners_ids = get_values(lst=['Meta', 'GA', 'Vungle'], p=[0.3, 0.2, 0.5], size=len(user_ids))
    views = get_values(lst=[1, 0], p=[0.1, 0.9], size=len(user_ids))
    
    
    df = pd.DataFrame(
    {'user_id':user_ids, 'country': countries, 
     'device': devices, 'partner_id':partners_ids, 
     'view':views}
    ) 
    
    print(f'shape{df.shape}')
    print(df.columns)
    
    return df
```
